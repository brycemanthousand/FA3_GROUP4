# 2. Suppose the cover price of a book is 24.95, but bookstores get a 40% discount. Shipping costs 3 for the first copy and 75 cents for each additional copy. What is the total wholesale cost for 60 copies?

```python
coverPrice = 24.95
discount = 0.4 
priceFirst = 3.00
priceAdditional = 0.75
copies = int(input("Enter number of copies: "))

bookDiscount = coverPrice * (1 - discount) * copies
shipping = priceAdditional * (copies - 1) + priceFirst
totalWholesaleCost = bookDiscount + shipping

print("The total wholesale cost for", copies, " copies is: $", round(totalWholesaleCost, 2)) 
```
